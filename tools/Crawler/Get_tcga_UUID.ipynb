```python
#get UUID
import requests
import json
import pandas as pd
from pandas import Series,DataFrame
import multiprocessing

def get_UUID(symbol,AA):
	web_head = "https://portal.gdc.cancer.gov/auth/api/v0/quick_search?query="
	web_tail = "&size=5"
	web_query = web_head + symbol + " " + AA + web_tail

	r = requests.get(web_query)
	t = json.loads(r.text)
	try:
		UUID = t['data']['query']['hits'][0]['ssm_id']
		
	except IndexError:
		UUID = "null"

	return UUID
#main
data = pd.read_csv('tcga_list.csv')
data = DataFrame(data)
length = len(data)
s = 0 
clock = 0
result =pd.DataFrame(columns=('symbol','aa_mutation','UUID'))
for index,row in data.iterrows():
	symbol = row['symbol']
	AA = row['aa_mutation']
	UUID = get_UUID(symbol,AA)
	s = s + 1
	clock = clock +1
	if clock == 50:
		print('(',s,'/',length,')')
		clock = 0
	result = result.append(pd.DataFrame({'symbol':[symbol],'aa_mutation':[AA],'UUID':[UUID]}),ignore_index=True)

result.to_csv("tcga_UUID_debug.csv",index=0)

```
